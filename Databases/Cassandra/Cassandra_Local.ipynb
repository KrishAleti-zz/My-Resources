- **Java Installation**

- **Apache Thrift Installation**
- **Cassandra Execution with Python**

#### Step no 1 - Download and Install Java: 
https://www.oracle.com/java/technologies/oracle-java-archive-downloads.html

1. https://www.oracle.com/java/technologies/javase/javase8-archive-downloads.html
2. https://www.oracle.com/webapps/redirect/signon?nexturl=https://download.oracle.com/otn/java/jdk/8u202-b08/1961070e4c9b4e26a04e7f5a083f551e/jdk-8u202-windows-x64.exe

Previous release [Link](https://www.oracle.com/java/technologies/oracle-java-archive-downloads.html). 

>Note: Only JRE is required on any 7 or 8 version of JAVA

Once installation is done, check for the JAVA folder created in Program Files of C drive, Copy the JAVA folder and paste in C drive.

To set the environment in local system, open Environment Variable:
> Note: Put name as JAVA_HOME and paste the `C drive path of java`


* In order to successfully verify the installation and PATH, run this given below command in command prompt
```cmd
java -version
```

* In order to verify the `Environment Variable` PATH, run this given below command in command prompt
```cmd
echo %JAVA_HOME%
```
    

    

#### Step no 2 - Download and Unzip Cassandra:

[Link](https://cassandra.apache.org/download/) for various version and operating system available to download.

https://www.apache.org/dyn/closer.lua/cassandra/3.11.12/apache-cassandra-3.11.12-bin.tar.gz

Once it is downloaded, extract it and copy the whole extracted file and paste it on `C` drive.
