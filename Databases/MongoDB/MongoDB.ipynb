##### $in operator:

```python
## Query documents using query operators($in,$lt,$gt)
# getting records in which the qualification is phd or master

for i in col.find({'qualification':{'$in':['phd','master']}}):
    print(i) 
```

##### And  operator:

```python
## And Query operator
# getting records in which the qualification is 'phd or master' and the age <35

for i in col.find({'qualification':'master','age':{'$lt':35}}):
    print(i)
```

```python
## And Query operator
# getting records in which the qualification is 'phd or master' and the age <35

for i in col.find({'$and':[{'qualification':'master'},{'age':{'$lt':35}}]}):
    print(i)
```

##### OR operator:

```python
## OR operators
# getting records in which the qualification is phd or first name is Ram

for i in col.find({'$or':[{'firstname':'Ram'},{'qualification':'phd'}]}):
    print(i)
```

#### $group

https://www.mongodb.com/docs/manual/reference/operator/aggregation/group/

```python
import pymongo 

# Connect with the portnumber and host  
client = pymongo.MongoClient("mongodb://localhost:27017/")  
    
# Access database  
mydatabase = client['Students2']  
    
# Access collection of the database  
collection=mydatabase['studentscores']  
data = [ 
    {"user":"Krish", "subject":"Database", "score":80}, 
    {"user":"Amit",  "subject":"JavaScript", "score":90}, 
    {"user":"Amit",  "title":"Database", "score":85}, 
    {"user":"Krish",  "title":"JavaScript", "score":75}, 
    {"user":"Amit",  "title":"Data Science", "score":60},
    {"user":"Krish",  "title":"Data Science", "score":95}] 
  
collection.insert_many(data) 
```

#### $sum

```python
# $group, $sum

#Find Amit and Krish's total subjects

#aggregate comes with grouby

agg_res= collection.aggregate( 
    [{ 
    "$group" :  
        {"_id" : "$user",   #we are grouping by 'user' 
         "Total records" : {"$sum" : 1}  #$sum  counts the records of each user
         }} 
    ])


for i in agg_res: 
    print(i)


#here total subject are fields that we want to get in output.
```

***IMP:***
In MongoDB, the $sum will add up the value of the expression which is 1 for every row and in our case, we are having 3 rows for our user so it will be 1+1+1 =3

https://stackoverflow.com/questions/40791907/what-does-sum1-mean-in-mongo

https://www.mongodb.com/docs/manual/reference/operator/aggregation/sum/

```python
### Calculating the total score based on user

agg_result= collection.aggregate( 
    [{ 
    "$group" :  
        {"_id" : "$user",  #grouping by user
         "Total Marks" : {"$sum" :"$score"} 
         }} 
    ]) 
for i in agg_result: 
    print(i)
```

#### $avg

https://www.mongodb.com/docs/manual/reference/operator/aggregation/avg/

```python
# $group, $avg

### Calculating the average score based on user

agg_result=collection.aggregate([
   {
      "$group": 
       {
         "_id": '$user',
         "StudentScoreAverage": {"$avg": "$score"}
       }
   }
])
for i in agg_result: 
    print(i)
```

```python
# $group, $avg, $multiply

### Calculating the average quantity And Average Price

agg_result=mycollection.aggregate([
   {
      "$group": {
         "_id": '$item', #grouping based on item
         "avgprice": {"$avg": {"$multiply": [ "$price", "$quantity" ]}},
          "avgQuantity": { "$avg": "$quantity" }
      }
   }
])
for i in agg_result: 
    print(i)
```

#### $Project

https://www.mongodb.com/docs/manual/reference/operator/aggregation/project/

```python
# $Project

import pymongo
client =pymongo.MongoClient("mongodb://localhost:27017/")

db=client['projectdb']
col=db['books']

data=[{
  "_id" : 1,
  "title": "abc123",
  "isbn": "0001122223334",
  "author": { "last": "zzz", "first": "aaa" },
  "copies": 5
},
{
  "_id" : 2,
  "title": "Baked Goods",
  "isbn": "9999999999999",
  "author": { "last": "xyz", "first": "abc", "middle": "" },
  "copies": 2
}
]


col.insert_many(data)
```

`$project` is just like the select statment, we need to give what fields are required in the output only those fields will be returned.

\<field>: <1 or true> --- Specifies the inclusion of a field. Non-zero integers are also treated as true.

We will get all the records/documenst but with the required fields (mentioned in the `$project` statement)


```python
for row in col.aggregate( [ { "$project": { "title": 1}} ] ):
    print(row)
```

```python
#SQL:- select title,isbn from col;

for row in col.aggregate( [ { "$project": { "title": 1,"isbn":1 }} ] ):
    print(row)
```

```python
for row in col.aggregate( [ { "$project": { "copies": 'true'}} ] ):
    print(row)
```
