```python
class test1:
    def __init__(self, a,b,c,d):
        self.a=a
        self.b=b
        self.c=c
        self.d=d
        
    def test_fn(self,v):
        return v-self.a

    def __str__(self):
        return "Hey there!!!"
    
class child_test1(test1):
    def __init__(self,x,a,b,c,d):
        super().__init__(a,b,c,d) #super()
        self.x=x
        
k=child_test1(12,1,8,7,8)
k.x
```

### Encapsulation:

```python
class test:
    def __init__(self, a,b,c):
        self.a =a 
        self.b = b
        self.c = c 
        
    def __str__(self):
        return "this is the return form my test class "

    
class test1 :
    def __init__(self, a,b,c):
        self.a =a 
        self.b = b
        self.c = c 
        
    def __str__(self):
        return "this is the return form my test1 class "
    
class test2:
    def __init__(self, a,b,c):
        self.a = a 
        self.b = b
        self.c = c 
        
    def __str__(self):
        return "this is the return form my test2 class "

class final:
    def __init__(self,x,y,z):
        self.x= x
        self.y= y
        self.z = z
        
    def __str__(self) : 
        return str(self.x) + "  " + str(self.y) + "  " + str(self.z)
```

```python
t= test(4,5,6)
t1 = test1(3,4,5)
t2 = test2(5,6,7)

f = final(t,t1,"sudh") #passed the class objects as arguments
print(f)
```

Here, we can also inherit the parent classes and call the functions/methods directly but the concept of Encapsulation is to hide the implementation.

So, we didn't inherit the classes and accessed. We didn' even know what methods or code does the test,test1,test2 classes have.

We had just passed their objects as arguments to our other class.
