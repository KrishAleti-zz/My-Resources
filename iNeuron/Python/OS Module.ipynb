```python
import os
# 1. create a directory
os.mkdir('dir1')
os.chdir(os.getcwd()+"\\dir1")
# 2. create 10 txt files and write date in all the files at a time
for i in range(10):
    f=open(f'os{i}.txt','w')
    f.write(f' fileno.{i}: create 10 txt files and write date in all the files at a time')
    f.close()
# 3. read all the data from this dir/files and print in console
# 4. hold all the data in the list
lst=[]
for i in os.listdir():
    file=open(i)
    print(file.read())
    file.seek(0)
    lst.append(file.read())
    file.close()
# 5. write all the data in a new file from the list.
file1=open("newf.txt",'w')
file1.writelines(lst)
file1.close()
# 6. keep only the new file in the present directory and move the other 10 files to some other dir.
os.chdir("..") #goes to one step back in the directory
os.mkdir(os.getcwd()+ '\\dir2')
import shutil
for i in range(10):
    shutil.move(f'{os.getcwd()}\\dir1\\os{i}.txt', f'{os.getcwd()}\\dir2\\os{i}.txt')
```
