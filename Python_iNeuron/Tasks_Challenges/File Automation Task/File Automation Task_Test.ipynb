Using tkinter for UI:

```python
#Import module
from tkinter import *

# Create object
root = Tk()
root.title('File Operations')

#set window color
root.configure(bg='black')

# Adjust size to Initialize tkinter window   
root.geometry("500x250")
#Need to give  max and minimum to keep the constant window size
# set minimum window size value
root.minsize(500, 250)
# set maximum window size value
root.maxsize(500, 250) 
 
    
# declaring string variable for storing name and password    
search_var=StringVar()

def searchfn():
    search=search_var.get()
    print("The Enter file name is : " + search)
    search_var.set("")
    
# creating a label for search using widget Label
search_label = Label(root, text = 'Enter the File Name: ', font=('calibre',10, 'normal'))
# creating a entry for searchusing widget Entry
search_entry = Entry(root, textvariable = search_var, font=('calibre',10,'normal'))

# creating a button using the widget
# Button that will call the search function
search_btn=Button(root,text = 'Search', font=('calibre',10, 'normal'), command = searchfn)

# placing the label and entry in the required position using grid method
search_label.grid(row=0,column=0)
search_entry.grid(row=0,column=1)
search_btn.grid(row=0,column=2)   
    
# performing an infinite loop for the window to display
root.mainloop()
```
