```python
!pip install python-docx
import docx #for working with docx files
import os  #for the directory and file management
import re  #regex for the match and search operations
import shutil  #In this code, to move the file to other dir

def for_foldername(string):
    """
    Returns the string before any numeric character in it after stipping off '_'
    
    Assignment_1 1.docx  --- Assignment, Assignment_1.docx --- Assignment, Assignment1.docx --- Assignment, 
    Programming_Assingment3.docx --- ProgrammingAssingment
    """
    for i in string:
        if i.isnumeric(): #checks if any numeric is there in the string(filename), if yes find the position of that numeric
            length = string.find(i)
            break
    return string[:length].replace('_','')

def rename_file(name):
    """
    This function renames the file by removing '(' and ')' from the file name.
    The source file name has to be passed as an argument and the function will 
    rename the file and return destination file name.
    
    Assingment_1(1).docx --- Assingment_1 1.docx
    """
    new_name = name.replace('(','').replace(')','')
    os.rename(name,new_name)
    return new_name


from docx import Document
mydoc = Document() # create an instance of a word document for creating a new document

"""getting only docx files and 
If there are any files having names ending with 1(1).docx,2(1).docx then changing files names to 1 1.docx and 2 2.docx"""
for i in os.listdir(): #searching all the files in the current dir
    ext=i.split(".")[-1] #splits the file name present in the dir and gives the last element of the splitted list
    if ext == 'docx': #after splitting if the last element is docx
        
        new_name = rename_file(i)

        # get new folder name by taking the file name prior to numeric part
        folder=for_foldername(new_name)
        
        #folder creation
        if not os.path.isdir(f'{os.getcwd()}\\{folder}'): 
            #we need to check whether required folder is already present or not, if not we will create it.
            os.mkdir(f'{os.getcwd()}\\{folder}')
        
        
        #adding the docx files and writing into the master file 
        mydoc.add_heading(i,2) # add a heading of level 2 (filename will be the heading)
        
        # create an instance of a word document we want to open
        doc = Document(new_name)
        
        for para in doc.paragraphs:
            mydoc.add_paragraph(para.text)
    #     mydoc.add_page_break()
        shutil.move(f'{os.getcwd()}\{new_name}',f'{os.getcwd()}\{folder}')
mydoc.save(f'{os.getcwd()}\Assignments_Masterfile.docx')
```
